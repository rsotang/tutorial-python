# Listas

Las listas son el tipo de datos nativo de python similar a los arrays

```python
#en las listas se puede guardar casi todo (incluido listas) y se pueden
#asignar a variables
#la funcion list() crea listas como si se tratase de un bucle de 
#llenado de arrays, recoge un argumento, asi que llenarlo de serie
#requiere llenarlo con una 

number_list=list((1,2,5))
print(number_list)

#la función range funciona igual que los argumentos de for, 
#creando un array uniformemente distribuido 

r=list(range(1,10))
print(r)

#ahora vamos a trata las funciones o métodos que podemos aplicar sobre 
#una lista

print(dir(number_list))

print(len(number_list)) #numero de elementos en la lista

#se pueden evaluar las listas de la misma forma que los arrays con 
#su orden de elementos

print(number_list[0])
print(7 in number_list) 

# in convierte en un condicional booleano la sentencia inicial con respecto la siguiente

#añade un elemento nuevo al final de la lista, 
#puede ser de cualquier tipo, solo uno
number_list.append(27)
print(number_list )

#para añadir elementos multiples al final tenemos que utilizar el método
#extend() que toma una lista como argumento, para separla y añadirla al final
number_list.extend((5,7))
print(number_list)

#para insertar elementos en una posicion concreta insert()
number_list.insert(1,1)
print(number_list)

#para quitar elementos del aray utilizamos el método pop() para lugares
#  y remove() para tipos de datos

number_list.pop(0)
print(number_list)

number_list.remove(5)
print(number_list)

#hay también métodos para ordenar cosa que en c es muy complicado de 
#implentar
#el método sort tiene parametros internos calibrables mediante bool
number_list.sort()
print(number_list)
number_list.sort(reverse=True)
print(number_list)

#podemos utilizar el mismo metodo que con strings de string slicing

number_list[1:]
number_list[:5]
number_list[2:4]

#para hacer listas vacias a rellenar utilizamos el comando list()
lista_nueva = list()

#en python tambien ha funciones preconstruidas que toman listas como argument

len(lista_nueva)
max(lista_nueva)
min(lista_nueva)
sum(lista_nueva)

#ejemplo de calculo con input con strings

while True:
    try:
            inp = input('mete un numero')
            if inp == 'done':
                break
            value = float(input)
            lista_nueva.append(value)
    except:
        print('no funciona')
        quit()
average = sum(lista_nueva)/len(lista_nueva)

print('Average:', average, 'Maximo:', max(lista_nueva))

#hay interacciones interesantes entre es el metodo split, que convierte
#una string en una lista ordenada basada en los espacios de serie
#si llenamos el argumento utiliza el llenado como delimitador para serparar


stringPrueba = 'una string de prueba'

cosa_nueva=stringPrueba.split() #esto ahora es una lista, split toma como argumento el caracter que toma como referencia para separar.

#con esto podemos implementar un algoritmo doble split para separar
#primero una linea, y despues palabras dentro de esa linea
```
