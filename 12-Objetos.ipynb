# Objetos

Primero vamos a especificar los términos de programación orientada a objetos los objetos son pequeños tipos de codigo intermedios entre funciones y programas.

Definiciones de conceptos

class : es el molde que define las caracteristicas de nuestro objeto

metodo: son los verbos de las clases, las modificaciones internas que se pueden hacer a la naturaleza del objeto

attributos: caracteristicas internas del objetos que se pueden modificar mediante parametros internos en la creación del objeto de tipo clase

objetos: es el caso particular de una clase predefinida
