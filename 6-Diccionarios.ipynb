# Diccionarios

Los diccionarios son un tipo de datos nativo de python que permite hacer asociaciones entre huecos de memoria de una forma clasificativa, es decir, es una pequeña base datos, o mecanismo de etiquetado de datos.

```python
#el diccionario es un set en el que una variable se le asignan caracteristicas específicas definibles que nosotros definimos tanto su tipo como su valor.

#las claves o keys siempre se especifican como strings mientras que los 

#items puede ser cualquier tipo de dato

producto={
    'name': 'libro',
    'quantity': 2,
    'price': 5.99

}

person = {
    'name': 'paco',
    'surname': 'illo'
}

#vamos a ver los métodos de los diccionarios

print(dir(producto))

print(producto.keys())

print(producto.items())

#el metodo .clear() vacia tanto las claves como los items

#los diccionarios como cualquier clase de dato se pueden introducir en listas 

listaDeLaCompra =   [
    {'product':'libro','cantidad':2,'precio':12},
    {'product':'cosas','cantidad':3,'precio':343}
]

# funcion dict() viene bien para crear diccionarios vacios que llenamos poco a poco

ejemplo = dict()

ejemplo['money']=12
ejemplo['tissues']= 24
ejemplo['casas']=134

print(ejemplo['money'])

ejemplo['tissues']= ejemplo['tissues']+2

#un ejemplo util de diccionarios es utilizarlos para crear histogramas, esto es util porque los histogramas se pueden modificar dinamicamente añadiendo categorias automaticamente segun salgan

#el operador in funciona con los diccionarios comprobando si hay keys de ese tipo en el diccionario

print('money' in ejemplo)

#de esta forma si una key no esta en el diccionario podemos añadirla de forma dinámica no preocuparnos por especificar las keys a la hora de hacer un conteo

counts = dict()

names = ['csev','cwen','csev','zquian','cwen']

for name in names:
    if name not in counts:
        counts[name]=1
    else:
        counts[name]=counts[name]+1
print(counts)

#el metodo get() hace todo esto de una vez, mira a ver si hay una key en el diccionario y si no la hay, la añade con un valor por defecto

#con esto podemos hacer un algoritmo de conteo simplificado

counts = dict()
names = ['csev','cwen','csev','zquian','cwen']
for name in names:
    counts[name]=counts.get(name,0)+1
print(counts)


```
